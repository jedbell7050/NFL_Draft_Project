```python
def create_player_tables(urls):
    cols = ['Player', 'Split', 'Value', 'Cmp', 'Att', 'Pct', 'Yds', 'TD', 'Int', 'Rate', 'Att.1', 'Yds.1', 'Avg', 'TD.1', 'Rec', 'Yds.2', 'Avg.1', 'TD.2', 'Solo', 'Ast', 'Tot', 'Loss', 'Sk', 'Int.1', 'Yds.3', 'TD.3', 'PD', 'FR', 'Yds.4', 'TD.4', 'FF']
    ndfs = pd.DataFrame(columns=cols)
    for url in urls:
        player_stat = pd.read_html(url, flavor='lxml', header=1)
        df = player_stat[0]

        data = df.iloc[0]

        ndf = pd.DataFrame(data)
        ndf.reset_index(inplace=True)
        ndf = ndf.transpose()
        ndf.columns = ndf.iloc[0]
        ndf.drop('index', inplace=True)

        for i in cols:
            if i not in ndf.columns:
                ndf[i] = np.nan


        pattern1 = r'http://www.sports-reference.com/cfb/players/'
        pattern2 = r'/splits/2019/'

        name = re.sub(pattern1, '', url)
        name = re.sub(pattern2, '', name)

        ndf['players'] = name

        ndf['players'] = ndf['players'].str.split('-')

        for index, item in ndf.iterrows():
            ndf['first'] = item['players'][0]
            ndf['last'] = item['players'][1]

        ndf['Player'] = ndf['first'].str.capitalize() +' ' + ndf['last'].str.capitalize()
        ndf['last'] = ndf['last'].str.capitalize()
        ndf['first'] = ndf['first'].str.capitalize()
        ndf['first_I'] = ndf['first'].apply(lambda x: x[0:2]).str.upper()
        ndf.drop(['players'], axis=1, inplace=True)
        
        r = requests.get(url)
        soup = BeautifulSoup(r.content, 'html.parser')

        school = soup.select('p a:last-of-type')
        school_info = []
        for lin in school:
            school_info.append(lin.get_text())

        college = re.sub('State', 'St.', school_info[0])
        ndf['College/Univ'] = college
        
        ndfs = pd.concat([ndfs, ndf])
    
    return ndfs
```

```python
from bs4 import BeautifulSoup
import requests

YEAR = np.arange(2010, 2021, 1)
all_coll_urls = []
for i in YEAR:
    url = 'https://www.pro-football-reference.com/years/'+str(i)+'/draft.htm'
    r = requests.get(url)

    page_soup = BeautifulSoup(r.content, 'html.parser')

    
    college_urls = []
    for line in page_soup.select('tbody tr'):
        for letter in line.select('td:nth-of-type(4)'):
            if (letter.get_text()!= 'T') and (letter.get_text()!= 'G') and (letter.get_text()!= 'C') and (letter.get_text()!= 'LS') and (letter.get_text()!= 'P') and (letter.get_text()!= 'K'):
                col_stats = line.select('td.right a')
                for stat in col_stats:
                    urls = stat.get('href')
                    urls = re.sub('.html', '/splits/'+str(i-1)+'/', urls)
                    college_urls.append(urls)
    all_coll_urls.append(college_urls)
        
flat_coll_urls = [item for sublist in all_coll_urls for item in sublist]
print(flat_coll_urls)

```

```python
def get_player_tables(urls):
    cols = ['Player', 'Split', 'Value', 'Cmp', 'Att', 'Pct', 'Yds', 'TD', 'Int', 'Rate', 'Att.1', 'Yds.1', 'Avg', 'TD.1', 'Rec', 'Yds.2', 'Avg.1', 'TD.2', 'Solo', 'Ast', 'Tot', 'Loss', 'Sk', 'Int.1', 'Yds.3', 'TD.3', 'PD', 'FR', 'Yds.4', 'TD.4', 'FF']
    ndfs = pd.DataFrame(columns=cols)
    for url in urls:
        try: 
            player_stat = pd.read_html(url, flavor='lxml', header=1)
        except ValueError:
            
            try:
                old_year = url[-5:-1]
                year = int(old_year)
                new_year = year-1
                new_year = str(new_year)
                new_url = re.sub(old_year, new_year, url)
                player_stat=pd.read_html(new_url, flavor='lxml', header=1)
                
            except ValueError:
                continue
                    
            
        df = player_stat[0]

        data = df.iloc[0]

        ndf = pd.DataFrame(data)
        ndf.reset_index(inplace=True)
        ndf = ndf.transpose()
        ndf.columns = ndf.iloc[0]
        ndf.drop('index', inplace=True)

        for i in cols:
            if i not in ndf.columns:
                ndf[i] = np.nan


        pattern1 = r'http://www.sports-reference.com/cfb/players/'
        pattern2 = r'/splits/\d\d\d\d/'

        name = re.sub(pattern1, '', url)
        name = re.sub(pattern2, '', name)

        ndf['players'] = name

        ndf['players'] = ndf['players'].str.split('-')

        for index, item in ndf.iterrows():
            ndf['first'] = item['players'][0]
            ndf['last'] = item['players'][1]

        ndf['Player'] = ndf['first'].str.capitalize() +' ' + ndf['last'].str.capitalize()
        ndf['last'] = ndf['last'].str.upper()
        ndf['first'] = ndf['first'].str.capitalize()
        ndf['first_I'] = ndf['first'].apply(lambda x: x[0:2]).str.upper()
        ndf.drop(['players'], axis=1, inplace=True)
        
        r = requests.get(url)
        soup = BeautifulSoup(r.content, 'html.parser')

        school = soup.select('p a:last-of-type')
        school_info = []
        for lin in school:
            school_info.append(lin.get_text())

        college = re.sub('State', 'St.', school_info[0])
        college = re.sub('Texas Christian', 'TCU', college)
        college = re.sub('UCF', 'Central Florida', college)
        college = re.sub('Pitt', 'Pittsburgh', college)
        college = re.sub('Ole Miss', 'Mississippi', college)
        
        ndf['College/Univ'] = college
        
        ndfs = pd.concat([ndfs, ndf])
    
    return ndfs
```
